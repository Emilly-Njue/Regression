```python
data['infield_type'] = data['infield_type'].fillna('Unknown')
data['infield_type'].value_counts()
```

```python
from sklearn.impute import SimpleImputer

left_field_imputed = data[['left_field_distance']]

#Instantiate a SimpleImputer with strategy="median"
imputer = SimpleImputer(strategy='median')

#Fit the imputer
imputer.fit(left_field_imputed)

#Transform left_field_imputed using the imputer and
# assign the result to leftfield_imputed
leftfield_imputed = imputer.transform(left_field_imputed)

leftfield_imputed
```

```python
data['left_field_distance'] = leftfield_imputed
data.head()
```

```python
center_field_imputed = data[['center_field_distance']]

imputer.fit(center_field_imputed)

centerfield_imputed = imputer.transform(center_field_imputed)

centerfield_imputed
```

```python
data['center_field_distance'] = centerfield_imputed
data.head()
```

```python
right_field_imputed = data[['right_field_distance']]

imputer.fit(right_field_imputed)

rightfield_imputed = imputer.transform(right_field_imputed)

rightfield_imputed
```

```python
data['right_field_distance'] = rightfield_imputed
data.head()
```

```python
drops = data[['backstop', 'fire_zone']]

data = data.drop(drops, axis=1)
```

## CONVERT CATEGORICAL COLUMNS INTO NUMERICAL

```python
data.info()
```

```python
print(data.has_lights.value_counts())
print(' ')
print(data.field_usage.value_counts())
print(' ')
print(data.infield_type.value_counts())

```

```python
from sklearn.preprocessing import OrdinalEncoder

lights = data[['has_lights']]

#Instantiate an OrdinalEncoder
encoder = OrdinalEncoder()

#Fit the encoder
encoder.fit(lights)

encoder.categories_
```

```python
#Transform using the encoder and assign the result
haslights = encoder.transform(lights).flatten()

haslights
```

```python
#Replace value of has_lights
data['has_lights'] = haslights
data.head()
```

## MULTIPLE CATEGORIES

```python
from sklearn.preprocessing import OneHotEncoder

cat_columns= ['field_usage', 'infield_type']

#Instantiate an OrdinalEncoder
ohe = OneHotEncoder(categories="auto", sparse=False, handle_unknown="ignore")

#fit encoder
ohe.fit(data[cat_columns])

#get new column names
new_cat_columns = ohe.get_feature_names(input_features=cat_columns)

#transform
columns = pd.DataFrame(ohe.fit_transform(data[cat_columns]), columns=new_cat_columns, index=data.index)

#replace the columns with transformed ones
data = pd.concat([data.drop(cat_columns, axis=1), columns], axis=1)
data.head(25)
```
